```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
data = [('Rwanda', 'Kigali'), ('Kenya', 'Nairobi'), ('Uganda', 'Kampala'), ('Tanzania', 'Dodoma')]
file_path = "/tmp/cicd_training_table"
df = spark.createDataFrame(data, ['Country', 'Capital'])
df.write.format('delta').mode('overwrite').save(file_path)
print("Data written to file path:", file_path)
```
